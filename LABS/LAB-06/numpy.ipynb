# BASIC NUMPY OPERATIONS

```python
import numpy as np

np.array([1,2,3]) + 0.5

```

```python
np.random.seed(1234)
A = np.random.randint(1, 10, (3,4))
B = np.random.randint(1, 10, (3,1))

A + B
```

```python
np.random.seed(4321)
A = np.random.randint(1, 10, (4,4))
B = np.random.randint(1, 10, (2,1))

A + B
#  will raise an error
```

```python
np.random.seed(1111)
A = np.random.randint(1, 10, (3,1,4))
B = np.random.randint(1, 10, (2,1))

A + B
```

```python
import numpy as np

A = np.array([3,11,4,5])
B = np.array([5,0,3])

A[:, np.newaxis] - B
```

# SOLUTIONS FOR NUMPY QUESTIONS IN LAB

```python
import numpy as np
import pandas as pd

arr = np.array([[1, 2, 3], [4, 5, 6]])
df = pd.DataFrame(arr, columns=["A", "B", "C"])
print(df)
```

```python
import numpy as np
import pandas as pd

arr = np.array([1, 2, 3, 4, 5])
df = pd.DataFrame({"values": arr})
df["cumsum"] = np.cumsum(arr)
print(df)
```

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```

```python
import numpy as np

arr = np.random.randint(1, 10, 15)
unique_values = np.unique(arr)
print(arr)
print(unique_values)
```

```python
import pandas as pd

df = pd.DataFrame([[1, 2], [3, 4]], columns=["A", "B"])
df_transposed = df.T
print(df_transposed)
```

```python
import numpy as np

arr1 = np.array([0, 10, 20, 40, 60])
arr2 = np.array([10, 30, 40])
common = np.intersect1d(arr1, arr2)
print(common)
```

```python
import numpy as np

arr1 = np.array([0, 10, 20, 40, 60, 80])
arr2 = np.array([10, 30, 40, 50, 70, 90])
diff = np.setdiff1d(arr1, arr2)
print(diff)
```

```python
import numpy as np

a = np.array([1, 2])
b = np.array([4, 5])
print(a > b)
print(a >= b)
print(a < b)
print(a <= b)
```
