```python
import pandas as pd
#part 1
```

```python
data = {
    'Name': ['Asha', 'Ravi', 'Kiran', 'Meena', 'John', 'Priya'],
    'Age': [20, 21, None, 22, 20, None],
    'Marks': [85, None, 78, 90, None, 88],
    'City': ['Hyderabad', 'Chennai', None, 'Mumbai', 'Delhi', 'Hyderabad']
}

df = pd.DataFrame(data)
```

```python
print("Original Dataset:")
print(df)
```

```python
print("\nMissing Values (True means missing):")
print(df.isnull())
```

```python
print("\nCount of Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)

print("\nDataset after filling missing Age with Mean:")
print(df)
```

```python
median_marks = df['Marks'].median()
df['Marks'].fillna(median_marks, inplace=True)

print("\nDataset after filling missing Marks with Median:")
print(df)
```

```python
mode_city = df['City'].mode()[0]
df['City'].fillna(mode_city, inplace=True)

print("\nDataset after filling missing City with Mode:")
print(df)
```

```python
# Recreate original dataset
df_original = pd.DataFrame(data)

# Drop rows where Marks are missing
df_dropped = df_original.dropna(subset=['Marks'])

print("\nDataset after dropping rows with missing Marks:")
print(df_dropped)
```

```python
print("\nOriginal Dataset:")
print(df_original)

print("\nCleaned Dataset (After Filling Missing Values):")
print(df)
```

```python
sales = [100, 120, 130, 115, 5000, 140, 125, 110]

df = pd.DataFrame(sales, columns=['Sales'])

print("Original Dataset:")
print(df)
```

```python
mean_before = df['Sales'].mean()
print("\nMean BEFORE Removing Outlier:", mean_before)
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
outliers = df[(df['Sales'] < lower_bound) | (df['Sales'] > upper_bound)]

print("\nDetected Outliers:")
print(outliers)
```

```python
df_clean = df[(df['Sales'] >= lower_bound) & (df['Sales'] <= upper_bound)]

print("\nDataset After Removing Outliers:")
print(df_clean)
```

```python
mean_after = df_clean['Sales'].mean()

print("\nMean BEFORE Removal:", mean_before)
print("Mean AFTER Removal:", mean_after)
```

```python
import numpy as np
marks = np.array([50, 60, 70, 80, 90])

print("Original Marks:", marks)
```

```python
min_val = np.min(marks)
max_val = np.max(marks)

min_max_normalized = (marks - min_val) / (max_val - min_val)

print("\nMin-Max Normalized Values:")
print(min_max_normalized)
```

```python
mean = np.mean(marks)
std = np.std(marks)

z_score_standardized = (marks - mean) / std

print("\nZ-score Standardized Values:")
print(z_score_standardized)
```

```python
print("\nComparison:")

print("\nOriginal Marks:", marks)
print("Min-Max Normalized:", min_max_normalized)
print("Z-score Standardized:", z_score_standardized)

print("\nRange of Min-Max:", min(min_max_normalized), "to", max(min_max_normalized))
print("Mean of Z-score (approx):", np.mean(z_score_standardized))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Hours_Studied': [2,4,6,8,10],
    'Marks': [40,50,65,80,95]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)
```

```python
correlation = df['Hours_Studied'].corr(df['Marks'])

print("\nCorrelation Coefficient:", correlation)
```

```python
if correlation > 0:
    print("Positive Correlation")
elif correlation < 0:
    print("Negative Correlation")
else:
    print("No Correlation")
```

```python
plt.scatter(df['Hours_Studied'], df['Marks'])
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Hours Studied vs Marks")
plt.show()
```

```python
if correlation > 0.7:
    print("Strong Positive Correlation")
elif correlation > 0.3:
    print("Moderate Positive Correlation")
elif correlation > 0:
    print("Weak Positive Correlation")
elif correlation < -0.7:
    print("Strong Negative Correlation")
elif correlation < -0.3:
    print("Moderate Negative Correlation")
else:
    print("Weak or No Correlation")
```

```python
data = {
    'Customer_ID': [1,2,3,4,5,2,6,7],
    'Age': [22,25,30,35,40,25,45,50],
    'Income': [25000,30000,None,40000,500000,30000,45000,None],
    'Gender': ['Male','Female','Male','Female','Male','Female','Male','Female'],
    'Purchase_Amount': [2000,3000,2500,2700,100000,3000,2900,2800],
    'City': ['Hyderabad','Chennai','Mumbai','Delhi','Mumbai','Chennai','Delhi','Hyderabad']
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)
```

```python
print("\nNull Values in Each Column:")
print(df.isnull().sum())
```

```python
print("\n replacing income values with medain")
income_median = df['Income'].median()
df['Income'].fillna(income_median , inplace = True )
print("\nAfter Filling Missing Income with Median:")
print(df)
```

```python
print("\n remove duplicates from customers ")
df = df.drop_duplicates(subset = ['Customer_ID'])
print("\nAfter Removing Duplicate Customers:")
print(df)
```

```python
Q1 = df['Purchase_Amount'].quantile(0.25)
Q3 = df['Purchase_Amount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Purchase_Amount'] < lower) | (df['Purchase_Amount'] > upper)]

print("\nOutliers in Purchase_Amount:")
print(outliers)
```

```python
df = df[(df['Purchase_Amount'] >= lower) & (df['Purchase_Amount'] <= upper)]

print("\nAfter Removing Outliers:")
print(df)
```

```python
mean_income = df['Income'].mean()
std_income = df['Income'].std()

df['Income_Standardized'] = (df['Income'] - mean_income) / std_income

print("\nAfter Standardization:")
print(df[['Income','Income_Standardized']])
```

```python
df['Gender_Encoded'] = df['Gender'].map({'Male':1, 'Female':0})

print("\nAfter Encoding Gender:")
print(df[['Gender','Gender_Encoded']])
```

```python
correlation = df['Income'].corr(df['Purchase_Amount'])

print("\nCorrelation between Income and Purchase Amount:", correlation)
```

```python
bins = [18,30,45,60]
labels = ['Young','Adult','Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

print("\nAge Bins:")
print(df[['Age','Age_Group']])
```

```python
df.to_csv("cleaned_customer_data.csv", index=False)

print("\nCleaned dataset saved as 'cleaned_customer_data.csv'")
```
